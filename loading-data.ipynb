```python
from geopy.geocoders import ArcGIS # importing the ArcGIS library for geocoding
nom = ArcGIS(scheme='http') # init var nom with ArcGIS(http scheme)
nom.geocode("3995 23rd St, San Francisco, CA 94114") # pass the complete address we wanna geocode
```

```python
# reading a csv file
import pandas
df6 = pandas.read_csv("supermarkets.csv")
df6
```

```python
# combining Address column with 3 other attributes
df6["Address"] = df6["Address"]+", "+df6["City"]+", "+df6["State"]+", "+df6["Country"]
df6
```

```python
# creating a new Coulmn called Coordinates
# using the .apply func to apply geocode lib func and get Coordinates for the addresses
df6["Coordinates"] = df6["Address"].apply(nom.geocode)
df6
```

```python
df6.Coordinates[0] # displaying the one of the coordinates row by itself
```

```python
# lambda funcs use to perform inbuilt function calls
# using lambda var l we copy lat and long values from coordinates and append them for each addresses
df6["Latitude"] = df6["Coordinates"].apply(lambda l: l.latitude)
df6["Longitude"] = df6["Coordinates"].apply(lambda l: l.longitude)
df6
```

```python
df6.drop(["City", "State", "Country", "Coordinates"], 1) # dropping some columns we don't need anymore
```
